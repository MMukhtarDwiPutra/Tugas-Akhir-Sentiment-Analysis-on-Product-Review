```python
import pandas as pd
import re
import nltk
import math

df = pd.read_csv("shopee.csv", sep='\s*;\s')
sentiWordNet = pd.read_csv('SentiWordNet_3.0.0.csv', sep='\t')
```

```python
df = df[899:]
df = df.reset_index()
```

```python
arrLabel = []
j = 0
for txt in df['text']:
    kalimat = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", txt).split())
    kalimat = re.sub(r'[.,]', ' ', kalimat)
    kalimat = kalimat.lower()
    tokenization = nltk.word_tokenize(kalimat)

    postag = nltk.pos_tag(tokenization)
    processWord = []
    isBut = False #Untuk pengecekan jika ada kata "but" dan "if" karena itu hanya perumpamaan, maka score = 0
    isWill = False
    for i in range (0, len(postag)): #proses untuk mencari apakah ada kata yang didepannya terdapat kata "not" agar nilainya bisa dibalik nanti
        if("JJR" == postag[i][1] or "VBG" == postag[i][1] or "JJ" == postag[i][1] and (not("first" == postag[i][0]) or not("few" == postag[i][0]) )or "VBN" == postag[i][1] or ("VB" == postag[i][1] and not("be" == postag[i][0])) or ("VBD" == postag[i][1] and not("was" == postag[i][0] or "been" == postag[i][0] or "had" == postag[i][0]))):
            if(isWill):
                isWill = False
            elif(not(isBut)):
                if(i != 0 and ("none" == postag[i-1][0] or "not" == postag[i-1][0] or "n't" == postag[i-1][0] or "nothing" == postag[i-1][0] or "no" == postag[i-1][0]) or ("none" == postag[i-2][0] or "not" == postag[i-2][0] or "n't" == postag[i-2][0] or "nothing" == postag[i-1][0])):
                    processWord += [[postag[i][0]]+[postag[i][1]]+["reverse"]]
                else:
                    processWord += [[postag[i][0]]+[postag[i][1]]+["no"]]
        elif("unfortunately" == postag[i][0] or "however" == postag[i][0] or "but" == postag[i][0] or "than" == postag[i][0] or ("that" == postag[i][0] and not("DT" == postag[i][1]))):
            if(len(processWord) != 0 ):
                processWord = processWord[:-1]
            else:
                isBut = False
        elif("will" == postag[i][0] or "unless" == postag[i][0]):
            isWill = True
            
        
    posScore = 0 #untuk menampung skor positif
    negScore = 0 #untuk menampung skor negatif
    for k in processWord:
        if(not(any(char.isdigit() for char in k[0]))): #Cek kadang POSTag salah, ada angka yang terdeteksi verb
            tmp = sentiWordNet[sentiWordNet['SynsetTerms'].str.contains(k[0])].to_numpy() #potong data agar lebih cepat prosesnya
        isCamus = False
        for i in range(len(tmp)): #Hitung score satu kata
            t = tmp[i][4].split(" ")
            word = [w.split("#")[0] for w in t]
            if(k[0] in word):
                tmpPosScore = tmp[i][2]
                tmpNegScore = tmp[i][3]
                if(k[2] == "reverse"):
                    tmpPosScore = tmp[i][3]
                    tmpNegScore = tmp[i][2]
                posScore = posScore + tmpPosScore
                negScore = negScore + tmpNegScore
                isCamus = True
                    
        if(len(tmp) == 0 or not isCamus):#Antisipasi kata yang tidak terdapat dalam kamus 
            posScore = 0
            negScore = 0
            break
        
    if(posScore > negScore):
        label = "1"
    elif(posScore < negScore):
        label = "0"
    else:
        label = "netral"
    arrLabel.append(label)
    print(posScore, negScore, processWord, j)
    j += 1
    
TP = 0
TN = 0
FP = 0
FN = 0

tesData = []
trainData = []
for i in range(len(arrLabel)):
    kalimat = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", df['text'][i]).split())
    tokenization = nltk.word_tokenize(kalimat)
    postag = nltk.pos_tag(tokenization)
    
    if(arrLabel[i] == "1" and df['label'][i] == 1):
        TP += 1
        trainData.append([df['text'][i],df['label'][i],df['domain'][i]])
    elif(arrLabel[i] == "0" and df['label'][i] == 0):
        TN += 1
        trainData.append([df['text'][i],df['label'][i],df['domain'][i]])
    elif(arrLabel[i] == "1" and df['label'][i] == 0):
        postag = [k for k in postag if "JJ" in k or "VBN" in k]
        print("FP ",i,postag)
        FP += 1
    elif(arrLabel[i] == "0" and df['label'][i] == 1):
        postag = [k for k in postag if "JJ" in k or "VBN" in k]
        print("FN ",i,postag)
        FN += 1
    if(arrLabel[i] == "netral"):
        tesData.append([df['text'][i],df['label'][i],df['domain'][i]])
        
tesData = pd.DataFrame(tesData, columns = ['text', 'label', 'domain']) 
trainData = pd.DataFrame(trainData, columns = ['text', 'label', 'domain']) 

acc = (TP + TN) / (TP + TN + FP + FN) * 100
prec = TP / (TP + FP) * 100
recall = TP / (TP + FN) * 100
f1 = 2*((prec*recall)/(prec+recall))
print("TP : ",TP)
print("TN : ",TN)
print("FP : ",FP)
print("FN : ",FN)
print(acc,"% acc")
print(prec,"% prec")
print(recall,"% recall")
print(f1,"% F1 Score")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import svm
from pprint import pprint
from sklearn.model_selection import RandomizedSearchCV
```

```python
X_train = trainData['text']
y_train = trainData['label']
ngram_range = (1,2)
min_df = 10
max_df = 1.
max_features = 300

tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=ngram_range,
                        stop_words=None,
                        lowercase=True,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)

features_train = tfidf.fit_transform(X_train).toarray()
labels_train = y_train
print(features_train.shape)
```

```python
classifier =svm.SVC(random_state=8)
random_grid = {'C': [.0001, .001, .01],
                  'kernel': ['linear', 'rbf', 'poly'],
                  'gamma': [.0001, .001, .01, .1, 1, 10, 100],
                  'degree': [1, 2, 3, 4, 5],
                  'probability': [True]
              }
               
random_search = RandomizedSearchCV(estimator=classifier,
                                   param_distributions=random_grid,
                                   n_iter=75,
                                   scoring='accuracy',
                                   cv=10, 
                                   verbose=2, 
                                   random_state=8)

random_search.fit(features_train, labels_train)

print("Hyper parameter terbaik : ")
print(random_search.best_params_)
print("")
print("Akurasi yang dihasilkan : ")
print(random_search.best_score_)
```

```python
best_classifier = random_search.best_estimator_
```

```python
def predictWithlexiconMethod(kal):
    tmp = []
    kalimat = kal.lower()
    tokenization = nltk.word_tokenize(kalimat)

    postag = nltk.pos_tag(tokenization)
    processWord = []
    isBut = False #Untuk pengecekan jika ada kata "but" dan "if" karena itu hanya perumpamaan, maka score = 0
    isWill = False
    for i in range (0, len(postag)): #proses untuk mencari apakah ada kata yang didepannya terdapat kata "not" agar nilainya bisa dibalik nanti
        if("JJR" == postag[i][1] or "VBG" == postag[i][1] or "JJ" == postag[i][1] and not("first" == postag[i][0]) or "VBN" == postag[i][1] or ("VB" == postag[i][1] and not("be" == postag[i][0])) or ("VBD" == postag[i][1] and not("was" == postag[i][0] or "been" == postag[i][0] or "had" == postag[i][0]))):
            if(isWill):
                isWill = False
            elif(not(isBut)):
                if(("not" == postag[i-1][0] or "n't" == postag[i-1][0] or "nothing" == postag[i-1][0] or "no" == postag[i-1][0]) or ("not" == postag[i-2][0] or "n't" == postag[i-2][0] or "nothing" == postag[i-1][0])):
                    processWord += [[postag[i][0]]+[postag[i][1]]+["reverse"]]
                else:
                    processWord += [[postag[i][0]]+[postag[i][1]]+["no"]]
        elif("unfortunately" == postag[i][0] or "but" == postag[i][0] or "than" == postag[i][0] or ("that" == postag[i][0] and not("DT" == postag[i][1]))):
            if(len(processWord) != 0 ):
                processWord = processWord[:-1]
            else:
                isBut = False
        elif("will" == postag[i][0] or "unless" == postag[i][0]):
            isWill = True
        
    posScore = 0 #untuk menampung skor positif
    negScore = 0 #untuk menampung skor negatif
    for k in processWord:
        kalimat = k[0]
        if(not(any(char.isdigit() for char in kalimat))): #Cek kadang POSTag salah, ada angka yang terdeteksi verb
            tmp = sentiWordNet[sentiWordNet['SynsetTerms'].str.contains(kalimat)].to_numpy() #potong data agar lebih cepat prosesnya
        isCamus = False
        for i in range(len(tmp)): #Hitung score satu kata
            t = tmp[i][4].split(" ")
            word = [w.split("#")[0] for w in t]
            if(kalimat in word):
                tmpPosScore = tmp[i][2]
                tmpNegScore = tmp[i][3]
                if(k[2] == "reverse"):
                    tmpPosScore = tmp[i][3]
                    tmpNegScore = tmp[i][2]
                posScore = posScore + tmpPosScore
                negScore = negScore + tmpNegScore
                isCamus = True      
        if(len(tmp) == 0 or not isCamus):#Antisipasi kata yang tidak terdapat dalam kamus 
            posScore = 0
            negScore = 0
            break
        
    if(posScore > negScore):
        return 1
    elif(posScore < negScore):
        return 0
    else:
        return "netral"
```

```python
def preProcessing(kal):
    kalimat = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", kal).split())
    kalimat = re.sub(r'[.,]', ' ', kalimat)
    kalimat = kalimat.lower()
    tokenization = nltk.word_tokenize(kalimat)
    filtered_sentence = [w for w in tokenization if not w in stop_words]
    string = ""
    for s in filtered_sentence:
        string += s+" "
    return string
```
